# **Pembangunan model bahasa unigram dan bigram**

**1. Import NLTK** untuk menggunakan fungsi **tokenisasi**

Perlu mengunduh package punkt untuk tokenisasi juga

```python
import nltk
nltk.download('punkt')
```

**2. Tokenisasi kalimat**

```python
sentence = "Because at the end of the day, I've always been, and will always, love you. "
tokens = nltk.word_tokenize(sentence)
```

**3.** Masukkan contoh paragraf yang terdiri atas beberapa kalimat, sebagai contoh teks yang akan dibangun model bahasanya

```python
par = "I love your smile as much as I love your cry. And I love your eyes as much as I love your lies. I love you from your curly hair, and all the way down to your little toes. I love each and every part of you. Your big nose, pouty lips, little ears, long legs, buffy arms, and chubby cheeks. You are perfect to me. If you were born in the ancient times, you'd be a myth. And I'd be your believer. "
print(par)
```

**4.** Ubah penulisan teks menjadi huruf kecil semua

```python
lc_par = par.lower()
print(lc_par)
```

**5.** Lakukan tokenisasi teks. Tokenisasi menjadi beberapa kalimat, kemudian menjadi beberapa token

```python
sent_text = nltk.sent_tokenize(lc_par)
print(sent_text)

mod_sentences = []
for sentence in sent_text:
  mod_sentence = []
  sent_tokens = nltk.word_tokenize(sentence)
  mod_sentence.append('<s>')
  for token in sent_tokens:
    mod_sentence.append(token)
  mod_sentence.append('</s>')
  mod_sentences.append(mod_sentence)
print(mod_sentence)
print(mod_sentences)
```

**6.** Masukkan data kata yang muncul pada teks beserta frekuensi kemunculannya ke dictionary freq_tab{}

```python
freq_tab = {}
total_count = 0
for sentence in mod_sentences:
  for token in sentence:
    if token in freq_tab:
      freq_tab[token] += 1
    else:
      freq_tab[token] = 1
    total_count += 1
print(freq_tab)
print(total_count)
```

**7. Buat model unigram:** hitung probability unigram, simpan dalam dictionary prob_tab{}

```python
prob_tab = {}
for token in freq_tab:
  prob_tab[token] = freq_tab[token]/total_count
print(prob_tab)
```

**8.** Tes hitung probabilitas sebuah kalimat berdasarkan model unigram

```python
test_sentence = "I love each and every part of you."
lc_test_sentence = test_sentence.lower()
test_tokens = nltk.word_tokenize(lc_test_sentence)
total_prob = 1.0

for test_token in test_tokens:
  total_prob = total_prob * prob_tab[test_token]
  print(test_token)
print(total_prob)
```

**9. Hitung frekuensi kemunculan bigram**, masukkan dalam freq_bigram_tab

```python
freq_bigram_tab = {}
for sentence in mod_sentences:
  for i in range(1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])
    if curr_bigram in freq_bigram_tab:
      freq_bigram_tab[curr_bigram] += 1
    else:
      freq_bigram_tab[curr_bigram] = 1
```

```python
print(freq_bigram_tab)
```

**10. Pembuatan model probability bigram**

```python
bigram_prob_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])
    if curr_bigram not in bigram_prob_tab:
      bigram_prob_tab[curr_bigram] = freq_bigram_tab[curr_bigram]/freq_tab[sentence[i-1]]
print(bigram_prob_tab)
```

**11.** Buat fungsi untuk mencari kata selanjutnya yang mempunyai bigram prob tertinggi berdasarkan input kata sebelumnya

```python
def get_max_bigram_prob(prev_word):
  max_prob_next_word = ''
  bigram_cand = {k: v for k, v in bigram_prob_tab.items() if k[0]==prev_word}
  if len(bigram_cand) > 0:
    bigram_cand = sorted(bigram_cand.items(), key=lambda x: x[1], reverse=True)
    k = bigram_cand[0]
    print('bigram dengan probability tertinggi = ' + str(k))
    max_prob_next_word = k[0][1]
  return max_prob_next_word
```

**12.** Tes prediksi kata selanjutnya

```python
prev_token = "lies"
predicted_next_word = get_max_bigram_prob(prev_token)
if len(predicted_next_word) > 0:
  print("prediksi kata selanjutnya setelah kata " + prev_token + ": " + predicted_next_word)
else:
  print("kata sebelum tidak ditemukan di vocabulary atau tidak ada bigram yang mengandung kata tsb sbg previous token")
```
